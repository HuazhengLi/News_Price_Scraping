```python
# get the name list of S&P 500 Companies
SP500 = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
SP500_name = SP500.Symbol.tolist()
```

```python
SP500
```

```python
# initialize the finnhub
finnhub_client = finnhub.Client(api_key = API_KEY)
```

```python
# set the start date and stop date of the news to download
start_date = '2021-02-10'
end_date = '2022-02-09'
```

```python
# scrape news from finnhub and save the news for one company as an individual csv file
for company in SP500_name:
    print(company)
    result = finnhub_client.company_news(company, _from = '2021-02-10', to = '2022-02-09')
    df = pd.DataFrame.from_dict(result)
    df.to_csv(path_or_buf = '/Users/lihuazheng/Desktop/Techlent_ML_Learning/Keystone_Project/News/{}.csv'.format(company), index = False, header = True)
    sleep(randint(2,3))
```

```python
# scrape the stock price for the above companies from yfinance
import yfinance as yf
stock_price = yf.download(SP500_name, start = start_date, end = end_date)['Adj Close'] # it has data that we do not need
```

```python
# calculate the price change percentage from previous day's price
stock_price_variation = stock_price.pct_change()
```
